```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
plt.plot(x, y)
plt.show()
```

```python
# 1st method
fig = plt.figure() # create a figure
ax = fig.add_subplot() # add some axes
plt.show()
```

```python
# 2nd method
fig = plt.figure() # create a figure
ax = fig.add_axes([1, 1, 1, 1]) # add some data
ax.plot(x, y)
plt.show()
```

```python
# 3rd method (recommended)
fig, ax = plt.subplots()
ax.plot(x, [50, 100, 200, 250]) # add some data
plt.show()
```

# Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4, 6]
y = [11, 22, 33, 44, 21]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,5))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple plot",
       xlabel="x-axis",
      ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
plt.show(1)
```
